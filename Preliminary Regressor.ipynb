```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./daily_lcd_good.csv")
pd.to_numeric(df['DailyPeakWindSpeed'])
df
```

```python
df.dtypes
```

```python
X = df.iloc[:-1, 1:-2]
y = df['Tommrow DailyAvegerageDryBulb']
y = y.iloc[:-1]
X.head(3)
```

```python
df = pd.concat([X, y], axis=1)
df
```

```python
(train, test) = train_test_split(df, train_size=0.8, shuffle=True)
```

```python
X_train = train.iloc[:-1, :-2]
y_train = train['Tommrow DailyAvegerageDryBulb']
X_test = test.iloc[:-1, :-2]
y_test = test['Tommrow DailyAvegerageDryBulb']
y_test = y_test.iloc[:-1]
y_train = y_train.iloc[:-1]
X_train
```

```python
# Creating a baseline using only the average temperature from the previous day
Xb_train = X_train['DailyAverageDryBulbTemperature']
yb_train = y_train
Xb_test = X_test['DailyAverageDryBulbTemperature']
yb_test = y_test
lrb = LinearRegression()
lrb.fit(Xb_train.values.reshape(-1, 1), yb_train)
lrb.score(Xb_test.values.reshape(-1, 1), yb_test)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```
