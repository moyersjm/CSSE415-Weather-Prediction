```python
df = pd.read_csv('weather_description.csv')
df_copy = df.copy()
df.head()
```

```python
for col in df.columns:
    df[col][df[col].str.contains('clear').fillna(False)] = 'clear'
    df[col][df[col].str.contains('clouds').fillna(False)] = 'cloudy'
    df[col][df[col].str.contains('fog').fillna(False)] = 'fog'
    df[col][df[col].str.contains('mist').fillna(False)] = 'fog'
    df[col][df[col].str.contains('haze').fillna(False)] = 'fog'
    df[col][df[col].str.contains('snow').fillna(False)] = 'snow'
    df[col][df[col].str.contains('sleet').fillna(False)] = 'snow'
    df[col][df[col].str.contains('thunder').fillna(False)] = None
    df[col][df[col].str.contains('rain').fillna(False)] = 'rain'
    df[col][df[col].str.contains('drizzle').fillna(False)] = 'rain'
    df[col][df[col].str.contains('smoke').fillna(False)] = None
    df[col][df[col].str.contains('dust').fillna(False)] = None
    df[col][df[col].str.contains('ash').fillna(False)] = None
```

```python
valid = True
i = 0
while valid:
    df.iloc[24*i:24*(i+1),1:] = df.iloc[24*i:24*(i+1),1:].mode().iloc[0,:]
    i = i + 1
    if 24*(i+1) > df.shape[0]:
        valid = False
```

```python
df.to_csv('events.csv')
```
