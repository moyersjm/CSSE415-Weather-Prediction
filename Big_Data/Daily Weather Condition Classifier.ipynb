```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./big_data_cond.csv')
df.head(3)
```

```python
train, test = train_test_split(df, test_size=0.1, random_state=1)
X_train = train.iloc[:, :-14]
y_train = train.iloc[:, -14:]
X_test = test.iloc[:, :-14]
y_test = test.iloc[:, -14:]
```

```python
scaler = StandardScaler()
X_train_num = X_train.select_dtypes(include='number')
X_train_cat = X_train.select_dtypes(exclude='number')
X_test_num = X_test.select_dtypes(include='number')
X_test_cat = X_test.select_dtypes(exclude='number')
```

```python
scaler.fit(X_train_num)
X_train_scaled = scaler.transform(X_train_num)
X_test_scaled = scaler.transform(X_test_num)
X_train_num = pd.DataFrame(X_train_scaled, index=X_train_num.index, columns=X_train_num.columns)
X_test_num = pd.DataFrame(X_test_scaled, index=X_test_num.index, columns=X_test_num.columns)
X_train_cat = pd.get_dummies(X_train_cat)
X_test_cat = pd.get_dummies(X_test_cat)
X_test_cat
```

```python
X_train = pd.concat([X_train_num, X_train_cat], axis=1)
X_test = pd.concat([X_test_num, X_test_cat], axis=1)
X_test
```

```python
def predict_day(day):
    idx = 14 - day
    rfCV.fit(X_train, y_train.iloc[:, idx])
    score = rfCV.best_estimator_.score(X_test, y_test.iloc[:, idx])
    print("Forecasting weather condition", day, "days ahead with 24 hours history:")
    print("Best Depth: ", rfCV.best_params_)
    print("Validation R^2", rfCV.best_score_)
    print("Test R^2", score)
    baseline_score = r2_score(y_test.iloc[:,idx],X_test.iloc[:,-24:].mode(axis=1))
    if score > baseline_score:
        print('\tGood!')
    else:
        print('\tBad!')
```

```python
grid = {'max_depth': range(10)}
rf = RandomForestClassifier()
rfCV = GridSearchCV(rf,param_grid=grid,n_jobs=-1)
```
