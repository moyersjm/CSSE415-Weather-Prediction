```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pd.read_csv("big_data_weather_desc.csv")
df.head(5)
```

```python
def daterange(start_date, end_date):
    delta = timedelta(hours=1)
    while start_date < end_date:
        yield start_date
        start_date += delta
```

```python
start = datetime(2012, 10, 1, 12, 00)
end = datetime(2017, 11, 30, 00, 00)

i = 0;
dates = []
for date in daterange(start, end):
    dates.append(date.strftime("%Y/%m/%d %H:%M"))
    i = i +1
```

```python
## Latitude
cols = list(df.columns)
lat = pd.DataFrame()
lat['datetime'] = dates
for col in cols:
    lat[col] = df[col].iloc[1]
lat = lat.drop('City', axis=1)
```

```python
## Longitude
cols = list(df.columns)
lng = pd.DataFrame()
lng['datetime'] = dates
for col in cols:
    lng[col] = df[col].iloc[2]
lng = lng.drop('City', axis=1)
```

```python
## elevation
cols = list(df.columns)
elv = pd.DataFrame()
elv['datetime'] = dates
for col in cols:
    elv[col] = df[col].iloc[3]
elv = elv.drop('City', axis=1)
```

```python
lat.to_csv("latitude.csv")
lng.to_csv('longitude.csv')
elv.to_csv('elevation.csv')
```
