```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('../DataSets/terre haute lcd.csv', low_memory=False)
```

```python
df.head(3)
```

```python
monthly = df[df.SOURCE == 6]
monthly = monthly.iloc[:, 21:40]
monthly
```

```python
daily = monthly[monthly.DailyAverageDryBulbTemperature.isnull() == False]
daily
```

```python
daily.to_csv('lcd_daily_cleaned.csv')
```

```python
daily = daily.drop('DailyAverageRelativeHumidity', axis=1)
daily = daily.drop('DailyAverageSeaLevelPressure', axis=1)
daily = daily.drop('DailyAverageWetBulbTemperature', axis=1)
daily
```

```python
daily['DailySnowDepth'] = daily['DailySnowDepth'].fillna(0)
daily['DailySnowfall'] = daily['DailySnowfall'].fillna(0) 
```

```python
daily
```

```python
tempy = daily.DailyAverageDryBulbTemperature
tempy = tempy.shift(-1)
tempy
# tempy = tempy.iloc[1:]
# tempy.loc[len(df.index)] = [0]
# tempy
```

```python
daily['Tommrow DailyAvegerageDryBulb'] = tempy
```

```python
daily["DailyPrecipitation"] = daily["DailyPrecipitation"].replace('T', 0)
daily = daily.replace('\*', '', regex=True)
daily["DailyPeakWindSpeed"] = daily["DailyPeakWindSpeed"].replace('s', '', regex=True)
daily["DailyPeakWindDirection"] = daily["DailyPeakWindDirection"].replace('s', '', regex=True)
daily
```

```python
daily.to_csv('daily_lcd_good.csv')
```
